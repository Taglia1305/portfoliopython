# Order Analysis with NumPy

**Project objective**  
This notebook analyzes a small order dataset using **NumPy**. The analysis calculates revenue, costs, margins, delivery performance, customer satisfaction, and a synthetic performance indicator called `quality_score`.

---

## Analysis structure

1. Dataset creation and inspection  
2. Extraction of the main variables  
3. Revenue, cost, and margin calculation  
4. Percentage margin analysis  
5. Order segmentation by delivery speed  
6. Construction of the `quality_score`  
7. Order classification  
8. Final report


## 1. Library import and dataset creation

In this section, the **NumPy** library is imported and the starting dataset is created.  
Each row represents one order, while each column contains a specific variable.

| Column | Variable | Meaning |
|---:|---|---|
| 0 | `quantita` | Number of units purchased |
| 1 | `prezzo_unitario` | Selling price per unit |
| 2 | `costo_unitario` | Unit cost |
| 3 | `giorni_consegna` | Order delivery time |
| 4 | `soddisfazione` | Customer satisfaction score |


```python
import numpy as np

dataset = np.array([
    [2, 50, 30, 3, 5],
    [1, 120, 80, 5, 4],
    [3, 40, 22, 2, 5],
    [5, 25, 15, 7, 3],
    [2, 200, 140, 4, 4],
    [1, 300, 210, 6, 3],
    [4, 35, 20, 2, 5],
    [2, 75, 45, 3, 4],
    [6, 20, 12, 8, 2],
    [1, 500, 350, 5, 5],
    [3, 60, 35, 4, 4],
    [2, 90, 55, 6, 3],
    [4, 45, 25, 3, 5],
    [1, 250, 180, 7, 2],
    [5, 30, 18, 2, 4],
    [2, 150, 100, 4, 5],
    [3, 80, 50, 5, 3],
    [1, 700, 500, 6, 4],
    [2, 110, 70, 3, 5],
    [4, 55, 32, 4, 4],
    [6, 18, 10, 9, 2],
    [3, 95, 60, 5, 3],
    [2, 180, 120, 4, 5],
    [1, 400, 280, 6, 4],
    [5, 22, 13, 7, 3],
    [2, 130, 85, 3, 5],
    [3, 70, 42, 4, 4],
    [4, 38, 21, 2, 5],
    [1, 600, 430, 8, 2],
    [2, 160, 105, 5, 4]
])
```

## 2. Preliminary dataset structure check

Before starting the analysis, it is useful to inspect a few technical characteristics of the array:

- number of dimensions;
- dataset shape;
- total number of elements;
- data type stored in the array.


```python
print(dataset.ndim)
print(dataset.shape)
print(dataset.size)
print(dataset.dtype)
```

## 3. Variable extraction

The dataset columns are separated into one-dimensional arrays.  
This step makes it easier to work with individual variables without repeatedly referencing column indexes.


```python
quantita = dataset[:,0]
prezzo_unitario = dataset[:,1]
costo_unitario = dataset[:,2]
giorni_consegna = dataset[:,3]
soddisfazione = dataset[:,4]
```

### Extracted variable check

Printing the arrays allows us to verify that each column has been selected correctly.


```python
print(quantita)
print(prezzo_unitario)
print(costo_unitario)
print(giorni_consegna)
print(soddisfazione)
```

## 4. Revenue, cost, and margin calculation per order

For each order, three key economic indicators are calculated:

- **revenue** = unit price × quantity;
- **cost** = unit cost × quantity;
- **margin** = revenue − cost.


```python
ricavo = prezzo_unitario * quantita
costo = costo_unitario * quantita
margine = ricavo - costo
```

### Economic values per individual order

This cell returns the arrays related to revenue, costs, and margins calculated for each transaction.


```python
print(f"Revenue for each individual transaction: {ricavo}")
print(f"Cost for each individual transaction: {costo}")
print(f"Margin for each individual transaction: {margine}")
```

## 5. Aggregate economic indicators

This section calculates the main aggregate metrics of the dataset:

- total revenue;
- total cost;
- total margin;
- average margin per order;
- order with the highest margin;
- order with the lowest margin.


```python
ricavo_totale = np.sum(ricavo)
costo_totale = np.sum(costo)
margine_totale = np.sum(margine)
margine_medio_ordine = np.mean(margine)
ordine_migliore = np.max(margine)
ordine_peggiore = np.min(margine)
```

### Aggregate indicator output

These values provide a first overall view of the economic performance of the orders.


```python
print(ricavo_totale)
print(costo_totale)
print(margine_totale)
print(margine_medio_ordine)
print(ordine_migliore)
print(ordine_peggiore)
```

## 6. Percentage margin

Percentage margin is used to evaluate the relative profitability of each order, regardless of its absolute value.

The formula used is:

$$
\text{Percentage margin} = \frac{\text{Margin}}{\text{Revenue}} \times 100
$$


```python
margine_percentuale = np.round(margine / ricavo * 100,2)
margine_percentuale_medio = np.round(np.mean(margine_percentuale),2)
```

### Percentage margin output

The first output shows the percentage margin of each order.  
The second output shows the average percentage margin of the dataset.


```python
print(margine_percentuale)
print(margine_percentuale_medio)
```

## 7. Identification of high-margin orders

A Boolean mask is created to select orders with a percentage margin greater than or equal to 40%.

This step introduces a core NumPy concept: **filtering through conditions**.


```python
maschera = margine_percentuale >= 40
margine_alto = margine_percentuale[maschera]
```

### Margins above the threshold

This output returns only the percentage margins that satisfy the defined condition.


```python
print(margine_alto)
```

### High-margin orders

The same mask is applied to the original dataset to retrieve the complete rows of high-margin orders.


```python
ordini_margine_alto = dataset[maschera]
print(ordini_margine_alto)
```

## 8. High-margin orders with fast delivery

Here, a double condition is applied:

- percentage margin greater than or equal to 40%;
- delivery within 3 days.

The goal is to identify the best orders from both an economic and operational perspective.


```python
margine_consegna = (margine_percentuale >= 40) & (giorni_consegna <= 3)
top_ordini = dataset[margine_consegna]
```

### Best operational orders output

These orders combine solid profitability with fast delivery times.


```python
print(top_ordini)
```

## 9. Percentage incidence of top orders

This section calculates how much the `top_ordini` subset weighs compared to the total number of orders in the dataset.


```python
numero_ordini = dataset.shape[0]
num_top_ordini = top_ordini.shape[0]
percentuale_top_ordini = np.round(num_top_ordini / numero_ordini * 100,2)
```

### Number and percentage of top orders

The percentage helps determine whether the best orders are isolated cases or a significant part of the dataset.


```python
print(numero_ordini)
print(num_top_ordini)
print(percentuale_top_ordini)
```

## 10. Delivery speed segmentation

Orders are divided into two groups:

- **fast delivery**: delivery within 3 days;
- **slow delivery**: delivery after more than 5 days.

This segmentation makes it possible to compare economic performance and customer satisfaction across two distinct operational categories.


```python
consegna_veloce = giorni_consegna <= 3
consegna_lenta = giorni_consegna > 5
ordini_consegna_veloce = dataset[consegna_veloce]
ordini_consegna_lenta = dataset[consegna_lenta]
```

### Fast and slow order output

The two outputs allow us to directly inspect the rows belonging to each group.


```python
print(ordini_consegna_veloce)
print(ordini_consegna_lenta)
```

## 11. Customer satisfaction comparison

This section compares overall average satisfaction with the average satisfaction of fast-delivery and slow-delivery orders.

The objective is to check whether faster delivery appears to be associated with higher customer satisfaction.


```python
soddisfazione_media = np.round(np.mean(soddisfazione),2)
soddisfazione_veloci = np.round(np.mean(ordini_consegna_veloce[:,4]),2)
soddisfazione_lenta = np.round(np.mean(ordini_consegna_lenta[:,4],),2)
```

### Average satisfaction output

The three outputs show, respectively:

1. overall average satisfaction;
2. average satisfaction for fast-delivery orders;
3. average satisfaction for slow-delivery orders.


```python
print(soddisfazione_media)
print(soddisfazione_veloci)
print(soddisfazione_lenta)
```

## 12. Economic comparison between fast and slow orders

In addition to customer satisfaction, average revenue and average margin are compared across the two groups.

This helps understand whether faster-delivered orders are also more attractive from an economic perspective.


```python
ricavo_medio_veloci = np.round(np.mean(ricavo[consegna_veloce]),2)
ricavo_medio_lenti = np.round(np.mean(ricavo[consegna_lenta]),2)
margine_medio_veloci = np.round(np.mean(margine[consegna_veloce]),2)
margine_medio_lenti = np.round(np.mean(margine[consegna_lenta]),2)
```

### Economic comparison output

The outputs show average revenue and average margin for fast and slow orders.


```python
print(ricavo_medio_veloci)
print(ricavo_medio_lenti)
print(margine_medio_veloci)
print(margine_medio_lenti)
```

## 13. Average percentage margin comparison

The comparison is completed by analyzing the average percentage margin for fast and slow orders.

This metric is useful because it normalizes profitability relative to the order value.


```python
margine_percentuale_medio_veloci = np.round(np.mean(margine_percentuale[consegna_veloce]),2)
margine_percentuale_medio_lenti = np.round(np.mean(margine_percentuale[consegna_lenta]),2)
```

### Average percentage margin by segment

The two outputs compare the relative profitability of fast and slow orders.


```python
print(margine_percentuale_medio_veloci)
print(margine_percentuale_medio_lenti)
```

## 14. Quality Score construction

The `quality_score` is a synthetic indicator created to evaluate each order by combining three dimensions:

- percentage margin;
- customer satisfaction;
- delivery days.

The formula rewards profitable orders with high customer satisfaction and penalizes longer delivery times.

$$
\text{Quality Score} = \text{Percentage margin} + (\text{Satisfaction} \times 10) - (\text{Delivery days} \times 2)
$$


```python
quality_score = np.round(margine_percentuale + soddisfazione * 10 - giorni_consegna * 2,2)
```

### Quality Score descriptive indicators

The average, maximum, and minimum values of the `quality_score` are calculated.


```python
quality_score_medio = np.round(np.mean(quality_score),2)
quality_score_massimo = np.round(np.max(quality_score),2)
quality_score_minimo = np.round(np.min(quality_score),2)
```

### Quality Score output

The first output shows the score for each order.  
The following outputs show the average, maximum, and minimum values.


```python
print(quality_score)
print(quality_score_medio)
print(quality_score_massimo)
print(quality_score_minimo)
```

## 15. Selection of high Quality Score orders

Orders with a `quality_score` greater than or equal to 80 are selected.

This threshold identifies the best orders according to the synthetic criterion defined above.


```python
maschera_qualità = quality_score >= 80
ordini_migliori_quality = dataset[maschera_qualità]
quality_alti = quality_score[maschera_qualità]
```

### Best orders by Quality Score

The first output shows the dataset rows.  
The second output shows the related scores.


```python
print(ordini_migliori_quality)
print(quality_alti)
```

## 16. Best order according to the Quality Score

Using `np.argmax()`, the index of the order with the highest score is identified.

> **Note:** from this point onward, the variable `ordine_migliore` refers to the best order according to the `quality_score`, not to the absolute economic margin calculated in the first sections.


```python
indice_migliore = np.argmax(quality_score)
ordine_migliore = dataset[indice_migliore]
quality_migliore = quality_score[indice_migliore]
```

### Best order output

The best order and its corresponding `quality_score` are printed.


```python
print(ordine_migliore)
print(quality_migliore)
```

## 17. Worst order according to the Quality Score

Using `np.argmin()`, the index of the order with the lowest score is identified.


```python
indice_peggiore = np.argmin(quality_score)
ordine_peggiore = dataset[indice_peggiore]
quality_peggiore = quality_score[indice_peggiore]
```

### Worst order output

The worst order and its corresponding `quality_score` are printed.


```python
print(ordine_peggiore)
print(quality_peggiore)
```

## 18. Qualitative order classification

Orders are classified into three categories:

| Class | Condition |
|---|---|
| `Excellent` | `quality_score >= 80` |
| `Good` | `quality_score >= 60` and below 80 |
| `Critical` | `quality_score < 60` |

The `np.where()` function makes it possible to create a vectorized conditional classification.


```python
classe_quality = np.where(quality_score >= 80, "Excellent",
                 np.where(quality_score >= 60, "Good", "Critical"))
print(classe_quality)
```

## 19. Count of qualitative classes

The number of orders belonging to each category is calculated.


```python
ordini_eccellenti = np.sum(classe_quality == "Excellent")
ordini_buoni = np.sum(classe_quality == "Good")
ordini_critici = np.sum(classe_quality == "Critical")
```

### Class count output

These outputs show how many orders are classified as excellent, good, or critical.


```python
print(ordini_eccellenti)
print(ordini_buoni)
print(ordini_critici)
```

## 20. Percentage distribution of classes

Using `np.unique()`, the existing classes and their respective occurrence counts are retrieved.  
Then, the percentage weight of each class over the total number of orders is calculated.


```python
classi, conteggi = np.unique(classe_quality, return_counts=True)
percentuali_ordini = np.round(conteggi / numero_ordini * 100,2)
print(classi)
print(conteggi)
print(percentuali_ordini)
```

## 21. Class summary table

Counts and percentages are combined into a single structure using `np.column_stack()`.


```python
riepilogativo = np.column_stack((conteggi, percentuali_ordini))
print(riepilogativo)
```

## 22. Sorting classes by frequency

Classes are sorted by the number of orders, from the most frequent to the least frequent.

This step makes the qualitative distribution easier to read.


```python
indici_ordinati = np.argsort(conteggi)[::-1]
classi_ordinati = classi[indici_ordinati]
conteggi_ordinati = conteggi[indici_ordinati]
percentuali_ordinati = percentuali_ordini[indici_ordinati]
```

### Sorted class output

The ordered indexes, classes, counts, and percentages are printed.


```python
print(indici_ordinati)
print(classi_ordinati)
print(conteggi_ordinati)
print(percentuali_ordinati)
```

## 23. Final report

The last section summarizes the main results of the analysis in a readable format.

The report includes:

- total revenue, costs, and margins;
- average percentage margin;
- average satisfaction;
- number of excellent and critical orders;
- best and worst order according to the `quality_score`.


```python
print("FINAL ORDER ANALYSIS REPORT")
print("-----------------------------")

print(f"Total revenue: {ricavo_totale}")
print(f"Total cost: {costo_totale}")
print(f"Total margin: {margine_totale}")
print(f"Average percentage margin: {margine_percentuale_medio}")
print(f"Average satisfaction: {soddisfazione_media}")

print(f"Excellent orders: {ordini_eccellenti}")
print(f"Critical orders: {ordini_critici}")

print(f"Best order: {ordine_migliore}")
print(f"Best order quality score: {quality_migliore}")

print(f"Worst order: {ordine_peggiore}")
print(f"Worst order quality score: {quality_peggiore}")
```
