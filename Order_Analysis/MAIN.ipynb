# Analisi ordini con NumPy

**Obiettivo del progetto**  
Questo notebook analizza un dataset di ordini utilizzando **NumPy**. L'analisi calcola ricavi, costi, margini, performance di consegna, soddisfazione cliente e un indicatore sintetico chiamato `quality_score`.

---

## Struttura dell'analisi

1. Creazione e lettura del dataset  
2. Separazione delle variabili principali  
3. Calcolo di ricavi, costi e margini  
4. Analisi dei margini percentuali  
5. Segmentazione degli ordini in base alla consegna  
6. Costruzione del `quality_score`  
7. Classificazione degli ordini  
8. Report finale


## 1. Importazione libreria e creazione del dataset

In questa sezione viene importata la libreria **NumPy** e viene costruito il dataset di partenza.  
Ogni riga rappresenta un ordine, mentre ogni colonna contiene una variabile specifica.

| Colonna | Variabile | Significato |
|---:|---|---|
| 0 | `quantita` | Numero di unità acquistate |
| 1 | `prezzo_unitario` | Prezzo di vendita per unità |
| 2 | `costo_unitario` | Costo sostenuto per unità |
| 3 | `giorni_consegna` | Tempo di consegna dell'ordine |
| 4 | `soddisfazione` | Valutazione del cliente |


```python
import numpy as np

dataset = np.array([
    [2, 50, 30, 3, 5],
    [1, 120, 80, 5, 4],
    [3, 40, 22, 2, 5],
    [5, 25, 15, 7, 3],
    [2, 200, 140, 4, 4],
    [1, 300, 210, 6, 3],
    [4, 35, 20, 2, 5],
    [2, 75, 45, 3, 4],
    [6, 20, 12, 8, 2],
    [1, 500, 350, 5, 5],
    [3, 60, 35, 4, 4],
    [2, 90, 55, 6, 3],
    [4, 45, 25, 3, 5],
    [1, 250, 180, 7, 2],
    [5, 30, 18, 2, 4],
    [2, 150, 100, 4, 5],
    [3, 80, 50, 5, 3],
    [1, 700, 500, 6, 4],
    [2, 110, 70, 3, 5],
    [4, 55, 32, 4, 4],
    [6, 18, 10, 9, 2],
    [3, 95, 60, 5, 3],
    [2, 180, 120, 4, 5],
    [1, 400, 280, 6, 4],
    [5, 22, 13, 7, 3],
    [2, 130, 85, 3, 5],
    [3, 70, 42, 4, 4],
    [4, 38, 21, 2, 5],
    [1, 600, 430, 8, 2],
    [2, 160, 105, 5, 4]
])
```

## 2. Controllo preliminare della struttura del dataset

Prima di procedere con l'analisi, è utile verificare alcune caratteristiche tecniche dell'array:

- numero di dimensioni;
- forma del dataset;
- numero totale di elementi;
- tipo di dato contenuto nell'array.


```python
print(dataset.ndim)
print(dataset.shape)
print(dataset.size)
print(dataset.dtype)
```

## 3. Estrazione delle variabili

Le colonne del dataset vengono separate in array monodimensionali.  
Questo passaggio rende più semplice lavorare con le singole variabili senza dover richiamare continuamente gli indici delle colonne.


```python
quantita = dataset[:,0]
prezzo_unitario = dataset[:,1]
costo_unitario = dataset[:,2]
giorni_consegna = dataset[:,3]
soddisfazione = dataset[:,4]
```

### Verifica delle variabili estratte

La stampa degli array permette di controllare che ogni colonna sia stata selezionata correttamente.


```python
print(quantita)
print(prezzo_unitario)
print(costo_unitario)
print(giorni_consegna)
print(soddisfazione)
```

## 4. Calcolo di ricavo, costo e margine per ordine

Per ogni ordine vengono calcolati tre indicatori economici fondamentali:

- **ricavo** = prezzo unitario × quantità;
- **costo** = costo unitario × quantità;
- **margine** = ricavo − costo.


```python
ricavo = prezzo_unitario * quantita
costo = costo_unitario * quantita
margine = ricavo - costo
```

### Output dei valori economici per singolo ordine

Questa cella restituisce gli array relativi a ricavi, costi e margini calcolati per ciascuna transazione.


```python
print(f"il ricavo per ogni singola transazione è {ricavo}")
print(f"il costo per ogni singola transazione è pari a {costo}")
print(f"il margine per ogni singola transazione è pari a {margine}")
```

## 5. Indicatori economici aggregati

In questa sezione vengono calcolate le principali metriche aggregate del dataset:

- ricavo totale;
- costo totale;
- margine totale;
- margine medio per ordine;
- ordine con margine massimo;
- ordine con margine minimo.


```python
ricavo_totale = np.sum(ricavo)
costo_totale = np.sum(costo)
margine_totale = np.sum(margine)
margine_medio_ordine = np.mean(margine)
ordine_migliore = np.max(margine)
ordine_peggiore = np.min(margine)
```

### Stampa degli indicatori aggregati

Questi valori offrono una prima lettura complessiva della performance economica degli ordini.


```python
print(ricavo_totale)
print(costo_totale)
print(margine_totale)
print(margine_medio_ordine)
print(ordine_migliore)
print(ordine_peggiore)
```

## 6. Margine percentuale

Il margine percentuale consente di valutare la redditività relativa di ciascun ordine, indipendentemente dal suo valore assoluto.

```python
margine_percentuale = np.round(margine / ricavo * 100,2)
margine_percentuale_medio = np.round(np.mean(margine_percentuale),2)
```

### Stampa dei margini percentuali

La prima stampa mostra il margine percentuale di ogni ordine.  
La seconda stampa mostra il margine percentuale medio del dataset.


```python
print(margine_percentuale)
print(margine_percentuale_medio)
```

## 7. Identificazione degli ordini ad alto margine

Viene creata una maschera booleana per selezionare gli ordini con margine percentuale maggiore o uguale al 40%.

Questo passaggio introduce un concetto centrale di NumPy: il **filtraggio tramite condizioni**.


```python
maschera = margine_percentuale >= 40
margine_alto = margine_percentuale[maschera]
```

### Margini superiori alla soglia

Questa stampa restituisce solo i margini percentuali che rispettano la condizione impostata.


```python
print(margine_alto)
```

### Ordini con margine alto

La stessa maschera viene applicata al dataset originale per recuperare le righe complete degli ordini ad alto margine.


```python
ordini_margine_alto = dataset[maschera]
print(ordini_margine_alto)
```

## 8. Ordini con margine alto e consegna veloce

Qui viene applicata una doppia condizione:

- margine percentuale almeno pari al 40%;
- consegna entro 3 giorni.

L'obiettivo è individuare gli ordini migliori sia dal punto di vista economico sia dal punto di vista operativo.


```python
margine_consegna = (margine_percentuale >= 40) & (giorni_consegna <= 3)
top_ordini = dataset[margine_consegna]
```

### Stampa dei migliori ordini operativi

Questi ordini combinano buona redditività e tempi di consegna rapidi.


```python
print(top_ordini)
```

## 9. Incidenza percentuale dei top ordini

Questa sezione calcola quanto pesano i `top_ordini` sul totale degli ordini presenti nel dataset.


```python
numero_ordini = dataset.shape[0]
num_top_ordini = top_ordini.shape[0]
percentuale_top_ordini = np.round(num_top_ordini / numero_ordini * 100,2)
```

### Numero e percentuale di top ordini

La percentuale aiuta a capire se gli ordini migliori sono casi isolati oppure una componente significativa del dataset.


```python
print(numero_ordini)
print(num_top_ordini)
print(percentuale_top_ordini)
```

## 10. Segmentazione per velocità di consegna

Gli ordini vengono divisi in due gruppi:

- **consegna veloce**: consegna entro 3 giorni;
- **consegna lenta**: consegna oltre 5 giorni.

Questa segmentazione permette di confrontare performance economiche e soddisfazione cliente tra due categorie operative distinte.


```python
consegna_veloce = giorni_consegna <= 3
consegna_lenta = giorni_consegna > 5
ordini_consegna_veloce = dataset[consegna_veloce]
ordini_consegna_lenta = dataset[consegna_lenta]
```

### Stampa degli ordini veloci e lenti

Le due stampe permettono di osservare direttamente le righe appartenenti ai due gruppi.


```python
print(ordini_consegna_veloce)
print(ordini_consegna_lenta)
```

## 11. Confronto della soddisfazione cliente

In questa sezione viene confrontata la soddisfazione media generale con quella degli ordini veloci e degli ordini lenti.

Questo passaggio serve a verificare se la rapidità di consegna sembra associarsi a una soddisfazione più alta.


```python
soddisfazione_media = np.round(np.mean(soddisfazione),2)
soddisfazione_veloci = np.round(np.mean(ordini_consegna_veloce[:,4]),2)
soddisfazione_lenta = np.round(np.mean(ordini_consegna_lenta[:,4],),2)
```

### Output della soddisfazione media

Le tre stampe mostrano rispettivamente:

1. soddisfazione media complessiva;
2. soddisfazione media degli ordini veloci;
3. soddisfazione media degli ordini lenti.


```python
print(soddisfazione_media)
print(soddisfazione_veloci)
print(soddisfazione_lenta)
```

## 12. Confronto economico tra ordini veloci e lenti

Oltre alla soddisfazione, vengono confrontati anche ricavi e margini medi dei due gruppi.

Questo permette di capire se gli ordini consegnati più rapidamente sono anche più interessanti dal punto di vista economico.


```python
ricavo_medio_veloci = np.round(np.mean(ricavo[consegna_veloce]),2)
ricavo_medio_lenti = np.round(np.mean(ricavo[consegna_lenta]),2)
margine_medio_veloci = np.round(np.mean(margine[consegna_veloce]),2)
margine_medio_lenti = np.round(np.mean(margine[consegna_lenta]),2)
```

### Output del confronto economico

Le stampe mostrano ricavo medio e margine medio per ordini veloci e lenti.


```python
print(ricavo_medio_veloci)
print(ricavo_medio_lenti)
print(margine_medio_veloci)
print(margine_medio_lenti)
```

## 13. Confronto del margine percentuale medio

Il confronto viene completato analizzando il margine percentuale medio per ordini veloci e ordini lenti.

Questa metrica è utile perché normalizza la redditività rispetto al valore dell'ordine.


```python
margine_percentuale_medio_veloci = np.round(np.mean(margine_percentuale[consegna_veloce]),2)
margine_percentuale_medio_lenti = np.round(np.mean(margine_percentuale[consegna_lenta]),2)
```

### Output del margine percentuale medio per segmento

Le due stampe confrontano la redditività relativa degli ordini veloci e lenti.


```python
print(margine_percentuale_medio_veloci)
print(margine_percentuale_medio_lenti)
```

## 14. Costruzione del Quality Score

Il `quality_score` è un indicatore sintetico creato per valutare ogni ordine combinando tre dimensioni:

- margine percentuale;
- soddisfazione cliente;
- giorni di consegna.

La formula premia gli ordini redditizi e con alta soddisfazione, penalizzando invece i tempi di consegna più lunghi.

$$
	ext{Quality Score} = 	ext{Margine percentuale} + (	ext{Soddisfazione} 	imes 10) - (	ext{Giorni consegna} 	imes 2)
$$


```python
quality_score = np.round(margine_percentuale + soddisfazione * 10 - giorni_consegna * 2,2)
```

### Indicatori descrittivi del Quality Score

Vengono calcolati valore medio, massimo e minimo del `quality_score`.


```python
quality_score_medio = np.round(np.mean(quality_score),2)
quality_score_massimo = np.round(np.max(quality_score),2)
quality_score_minimo = np.round(np.min(quality_score),2)
```

### Stampa del Quality Score

La prima stampa mostra il punteggio di ogni ordine.  
Le stampe successive mostrano media, massimo e minimo.


```python
print(quality_score)
print(quality_score_medio)
print(quality_score_massimo)
print(quality_score_minimo)
```

## 15. Selezione degli ordini con Quality Score elevato

Vengono selezionati gli ordini con `quality_score` maggiore o uguale a 80.

Questa soglia identifica gli ordini considerati migliori secondo il criterio sintetico definito.


```python
maschera_qualità = quality_score >= 80
ordini_migliori_quality = dataset[maschera_qualità]
quality_alti = quality_score[maschera_qualità]
```

### Stampa degli ordini migliori per Quality Score

La prima stampa mostra le righe del dataset.  
La seconda stampa mostra i relativi punteggi.


```python
print(ordini_migliori_quality)
print(quality_alti)
```

## 16. Miglior ordine secondo il Quality Score

Con `np.argmax()` viene individuato l'indice dell'ordine con il punteggio più alto.

> **Nota:** da questo punto in poi la variabile `ordine_migliore` si riferisce al miglior ordine secondo il `quality_score`, non più al margine economico assoluto calcolato nelle prime sezioni.


```python
indice_migliore = np.argmax(quality_score)
ordine_migliore = dataset[indice_migliore]
quality_migliore = quality_score[indice_migliore]
```

### Output del miglior ordine

Vengono stampati l'ordine migliore e il relativo `quality_score`.


```python
print(ordine_migliore)
print(quality_migliore)
```

## 17. Peggior ordine secondo il Quality Score

Con `np.argmin()` viene individuato l'indice dell'ordine con il punteggio più basso.


```python
indice_peggiore = np.argmin(quality_score)
ordine_peggiore = dataset[indice_peggiore]
quality_peggiore = quality_score[indice_peggiore]
```

### Output del peggior ordine

Vengono stampati l'ordine peggiore e il relativo `quality_score`.


```python
print(ordine_peggiore)
print(quality_peggiore)
```

## 18. Classificazione qualitativa degli ordini

Gli ordini vengono classificati in tre categorie:

| Classe | Condizione |
|---|---|
| `Eccellente` | `quality_score >= 80` |
| `Buona` | `quality_score >= 60` e inferiore a 80 |
| `Critico` | `quality_score < 60` |

La funzione `np.where()` permette di creare una classificazione condizionale vettorializzata.


```python
classe_quality = np.where(quality_score >= 80, "Eccellente",
                 np.where(quality_score >= 60, "Buona", "Critico"))
print(classe_quality)
```

## 19. Conteggio delle classi qualitative

Viene calcolato il numero di ordini appartenenti a ciascuna categoria.


```python
ordini_eccellenti = np.sum(classe_quality == "Eccellente")
ordini_buoni = np.sum(classe_quality == "Buona")
ordini_critici = np.sum(classe_quality == "Critico")
```

### Output del conteggio classi

Queste stampe mostrano quanti ordini sono classificati come eccellenti, buoni o critici.


```python
print(ordini_eccellenti)
print(ordini_buoni)
print(ordini_critici)
```

## 20. Distribuzione percentuale delle classi

Con `np.unique()` vengono recuperate le classi presenti e il rispettivo numero di occorrenze.  
Successivamente viene calcolata la percentuale di ogni classe sul totale degli ordini.


```python
classi, conteggi = np.unique(classe_quality, return_counts=True)
percentuali_ordini = np.round(conteggi / numero_ordini * 100,2)
print(classi)
print(conteggi)
print(percentuali_ordini)
```

## 21. Tabella riepilogativa classi

I conteggi e le percentuali vengono uniti in un'unica struttura tramite `np.column_stack()`.


```python
riepilogativo = np.column_stack((conteggi, percentuali_ordini))
print(riepilogativo)
```

## 22. Ordinamento delle classi per frequenza

Le classi vengono ordinate in base al numero di ordini, dalla più frequente alla meno frequente.

Questo passaggio rende più immediata la lettura della distribuzione qualitativa.


```python
indici_ordinati = np.argsort(conteggi)[::-1]
classi_ordinati = classi[indici_ordinati]
conteggi_ordinati = conteggi[indici_ordinati]
percentuali_ordinati = percentuali_ordini[indici_ordinati]
```

### Output delle classi ordinate

Vengono stampati gli indici ordinati, le classi, i conteggi e le percentuali ordinate.


```python
print(indici_ordinati)
print(classi_ordinati)
print(conteggi_ordinati)
print(percentuali_ordinati)
```

## 23. Report finale

L'ultima sezione sintetizza i principali risultati dell'analisi in un formato leggibile.

Il report include:

- ricavi, costi e margini totali;
- margine percentuale medio;
- soddisfazione media;
- numero di ordini eccellenti e critici;
- miglior e peggior ordine secondo il `quality_score`.


```python
print("REPORT FINALE ANALISI ORDINI")
print("-----------------------------")

print(f"Ricavo totale: {ricavo_totale}")
print(f"Costo totale: {costo_totale}")
print(f"Margine totale: {margine_totale}")
print(f"Margine percentuale medio: {margine_percentuale_medio}")
print(f"Soddisfazione media: {soddisfazione_media}")

print(f"Ordini eccellenti: {ordini_eccellenti}")
print(f"Ordini critici: {ordini_critici}")

print(f"Miglior ordine: {ordine_migliore}")
print(f"Quality score miglior ordine: {quality_migliore}")

print(f"Peggior ordine: {ordine_peggiore}")
print(f"Quality score peggior ordine: {quality_peggiore}")
```
