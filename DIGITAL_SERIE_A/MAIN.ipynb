```python
def calculate_points(selected_team):
    points = 0
    for match in li:
        home_team = match[1]
        away_team = match[2]
        home_goals = match[3]
        away_goals = match[4]

        if home_team == selected_team:
            if home_goals > away_goals:
                points += 3
            elif away_goals == home_goals:
                points += 1

        if away_team == selected_team:
            if away_goals > home_goals:
                points += 3
            elif away_goals == home_goals:
                points += 1
    return points
```

```python
def team_stats(selected_team):
    w = 0 # wins
    d = 0 # draws
    l = 0 # losses
    for match in li:
        home_team = match[1]
        away_team = match[2]
        home_goals = match[3]
        away_goals = match[4]

        if home_team == selected_team:
            if home_goals > away_goals:
                w += 1
            elif home_goals == away_goals:
                d += 1
            elif home_goals < away_goals:
                l += 1

        if away_team == selected_team:
            if away_goals > home_goals:
                w += 1
            elif away_goals == home_goals:
                d += 1
            elif away_goals < home_goals:
                l += 1
    return w, d, l
```

```python
def final_standings():
    standings = {}

    for match in li:
        team = match[1]

        if team not in standings:
            standings[team] = calculate_points(team)
    return standings
```

```python
# --- MAIN INTERACTIVE LOOP ---

while True:
    print("\n--- SERIE A 08/09 DIGITAL STANDINGS ---")
    print("Enter a number from 1 to 3 and follow the on-screen instructions")
    print("1: Calculate total points for a specific team (use capital initial)")
    print("2: Get full statistics for a specific team (use capital initial)")
    print("3: View the complete Serie A 08/09 league table")

    choice = input("Enter a number from one to three (or 'exit'): ").lower()

    if choice == "exit":
        print("Thank you for using Digital Serie A. See you soon!")
        break

    elif choice == "1":
        t = input("Enter team name to calculate points: ")
        print(f"Total points: {calculate_points(t)}")

    elif choice == "2":
        t = input("Enter team name for statistics: ")
        wins, draws, losses = team_stats(t)
        print(f"Wins = {wins}, Losses = {losses}, Draws = {draws}")

    elif choice == "3":
        results = final_standings()
        print("\n--- FINAL LEAGUE TABLE ---")
        print(results)

    else:
        print("Invalid option, please try again.")
```
