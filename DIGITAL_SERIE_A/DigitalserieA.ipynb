```python
def calcola_punti(squadra_scelta):
  punti = 0
  for partita in li:
    squadra_casa = partita[1]
    squadra_trasferta = partita[2]
    goal_casa = partita[3]
    goal_trasferta = partita[4]

    if squadra_casa == squadra_scelta:
      if goal_casa > goal_trasferta:
        punti += 3
      elif goal_trasferta == goal_casa:
        punti += 1

    if squadra_trasferta == squadra_scelta:
      if goal_trasferta > goal_casa:
        punti += 3
      elif goal_trasferta == goal_casa:
        punti += 1
  return punti


```

```python
def stat_squadra(squadra_scelta):
  v = 0
  p = 0
  s = 0
  for partita in li:
    squadra_casa = partita[1]
    squadra_trasferta = partita[2]
    goal_casa = partita[3]
    goal_trasferta = partita[4]

    if squadra_casa == squadra_scelta:
      if goal_casa > goal_trasferta:
        v += 1
      elif goal_casa == goal_trasferta:
        p += 1
      elif goal_casa < goal_trasferta:
        s += 1

    if squadra_trasferta == squadra_scelta:
      if goal_trasferta > goal_casa:
        v += 1
      elif goal_trasferta == goal_casa:
        p += 1
      elif goal_trasferta < goal_casa:
        s += 1
  return v,p,s
```

```python
def classifica_finale():
  classifica = {}

  for partita in li:
    squadra = partita[1]

    if squadra not in classifica:
      classifica[squadra] = calcola_punti(squadra)
  return classifica
```
