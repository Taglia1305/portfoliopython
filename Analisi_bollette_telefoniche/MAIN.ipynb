```python
def converti_in_minuti(data):
  parti = data.split(":")

  giorno = int(parti[0])
  ora = int(parti[1])
  minuto = int(parti[2])

  if not (1 <= giorno <= 31):
    return -1
  if not(1 <= ora <= 23):
    return -1
  if not(1<= minuto <= 59):
    return -1

  minuti_totali = (giorno - 1) * 24 * 60 + ora * 60 + minuto

  return minuti_totali
```

```python
def ottieniDatiTelefonate(nomeFile):
  lista_risultato = []

  file = open(nomeFile,"r")
  file.readline()

  for riga in file:
    riga = riga.strip()

    if riga == "":
      continue
    if riga.startswith("#"):
      continue
    parti = riga.split(";")

    Cod_Chiamante = int(parti[0])
    Cod_Destinatario = int(parti[1])
    Cod_Cella_Chiamante = int(parti[2])
    Cod_Cella_Destinatario = int(parti[3])

    if Cod_Chiamante == Cod_Destinatario:
      continue

    inizio_totale = converti_in_minuti(parti[4])
    fine_totale = converti_in_minuti(parti[5])

    if inizio_totale == -1:
      continue
    if fine_totale == -1:
      continue

    if inizio_totale > fine_totale:
      continue

    durata = fine_totale - inizio_totale + 1

    if durata >= 600:
      continue

    telefonata_valida = (Cod_Chiamante, Cod_Destinatario, Cod_Cella_Chiamante, Cod_Cella_Destinatario, durata)
    lista_risultato.append(telefonata_valida)

  file.close()
  return lista_risultato
```

```python
def calcolaBollette(datiChiamate):
    minuti_totali = {} # Dizionario per accumulare i minuti

    for tel in datiChiamate:
        chiama = tel[0]
        durata = tel[4]
        # Qui accumula i minuti nel dizionario come abbiamo visto prima...
        if chiama not in minuti_totali:
            minuti_totali[chiama] = durata
        else:
            minuti_totali[chiama] += durata

    bollette = {}
    for cliente in minuti_totali:
        minuti = minuti_totali[cliente]

        # Applichiamo il bonus dei 20 minuti
        minuti_da_pagare = max(0, minuti - 20)

        bollette[cliente] = minuti_da_pagare * 0.10

    return bollette
```

```python
def calcolaCelleCongestionate(datiChiamate):
  aree_traffico = {}

  for area in datiChiamate:
    zona = area[2]

    if zona not in aree_traffico:
      # Se è la prima volta che vediamo questa cella, mettiamo 1
      aree_traffico[zona] = 1
    else:
    # Se la cella esiste già, aggiungiamo 1 al conteggio precedente
      aree_traffico[zona] += 1

  totale_chiamate = sum(aree_traffico.values())
  numero_celle = len(aree_traffico)
  if numero_celle == 0:
    return []

  media = totale_chiamate / numero_celle

  celle_congestionate = []
  for cod_cella in aree_traffico:
    quante_chiamate = aree_traffico[cod_cella]

    if aree_traffico[cod_cella] > media:
      celle_congestionate.append(cod_cella)

  return celle_congestionate
```

```python
# 1. Trasformi il file in una lista pulita
dati = ottieniDatiTelefonate("telefonate.csv")

# 2. Usi quella lista per le bollette
bollette = calcolaBollette(dati)

# 3. Usi la stessa lista per le celle
celle_critiche = calcolaCelleCongestionate(dati)

print("Bollette calcolate:", bollette)
print("Celle sopra la media:", celle_critiche)
```
