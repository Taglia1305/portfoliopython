```python
book_list = {}

# add_book(title, copies)
# Description: Adds a new book to the library with the specified number of copies.
# If the book already exists, it updates the copy count by adding the new ones.
# If the book does not exist, it adds it to the system.

def add_book(title, copies):
    if title not in book_list:
        book_list[title] = copies
    else:
        book_list[title] += copies

# remove_book(title)
# Description: Removes a book from the library.
# If the book exists, it is deleted from the system.
# If the book does not exist, the program prints an error message.

def remove_book(title):
    if title in book_list:
        del book_list[title]
    else:
        print("Error: book not found")

# check_availability(title)
# Description: Checks if a book is available in the library.
# Returns True if at least one copy is available.
# Returns False if the book does not exist or no copies are available.

def check_availability(title):
    if title in book_list:
        if book_list[title] >= 1:
            return True, f"remaining copies: {book_list[title]}"
        else:
            return False
    else:
        return False

# borrow_book(title)
# Description: Reduces the number of available copies of a book by 1, simulating a loan.
# If the book is available, it decrements the copy count by 1.
# If no copies are available or the book does not exist, it prints an error message.

def borrow_book(title):
    if title in book_list:
        if book_list[title] >= 1:
            book_list[title] -= 1
            print(f"Loan registered: {title}, remaining copies {book_list[title]}")
        else:
            print("Insufficient copies")
    else:
        print("Book not found")

# library_statistics()
# Description: Returns a dictionary with library information.
# total_titles: total number of unique titles.
# total_copies: total number of all copies.
# average_copies: average number of copies per book.

def library_statistics():
    n_titles = len(book_list)
    if n_titles == 0:
        return {
            "n_titles": 0,
            "total_copies": 0,
            "avg_copies": 0
        }
    total_copies = sum(book_list.values())
    avg_copies = total_copies / n_titles

    statistics = {
        "n_titles": n_titles,
        "total_copies": total_copies,
        "avg_copies": avg_copies
    }
    return statistics

# display_books()
# Description: Shows a list of all books in the library along with the number of available copies.
# If the library is empty, it prints a message indicating no books are present.

def display_books():
    if not book_list:
        print("The library is empty")
    else:
        for title, copies in book_list.items():
            print(f"Title: {title} | Quantity: {copies}")

# restore_book(title, copies)
# Description: Restores (adds) copies of a book that was borrowed or needs more stock.
# If the book exists, it adds the specified number of copies.
# If the book does not exist, it prints an error message.

def restore_book(title, copies):
    if title in book_list:
        book_list[title] += copies
    else:
        print("The requested book does not exist")

# --- MAIN INTERACTIVE LOOP ---

while True:
    print("\n--- WELCOME TO BIBLIOSOFTWARE ---")
    print("1. Add new title or copies")
    print("2. Remove a title from the system")
    print("3. Check availability")
    print("4. Borrow a book (-1 copy)")
    print("5. Restore/Return copies")
    print("6. View statistics")
    print("7. View all books")
    print("exit. Close the program")

    choice = input("\nWhat would you like to do? ").lower()

    if choice == "exit":
        print("Thank you for using BiblioSoftware. See you soon!")
        break

    elif choice == "1":
        t = input("Book title: ")
        c = int(input("Number of copies: "))
        add_book(t, c)

    elif choice == "2":
        t = input("Title to remove: ")
        remove_book(t)

    elif choice == "3":
        t = input("Title to search: ")
        is_available = check_availability(t)
        if is_available:
            print("Available!")
        else:
            print("Not available.")

    elif choice == "4":
        t = input("Which book would you like to borrow? ")
        borrow_book(t)

    elif choice == "5":
        t = input("Which book did you return or restore? ")
        c = int(input("How many copies? "))
        restore_book(t, c)

    elif choice == "6":
        stats = library_statistics()
        print("\n--- STATISTICS ---")
        print(f"Total titles: {stats['n_titles']}")
        print(f"Total copies: {stats['total_copies']}")
        print(f"Average copies: {stats['avg_copies']:.2f}")

    elif choice == "7":
        display_books()

    else:
        print("Invalid option, please try again.")
```
