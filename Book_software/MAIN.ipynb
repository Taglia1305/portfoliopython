```python
lista_libri = {}
```

#aggiungi_libro(titolo, copie)
Descrizione: Aggiunge un nuovo libro alla biblioteca con il numero di copie specificato.
Se il libro esiste già, aggiorna il numero di copie aggiungendo quelle nuove.
Se il libro non esiste, lo aggiunge al sistema.


```python
def aggiungi_libro(titolo, copie):

  if titolo not in lista_libri:
    lista_libri[titolo] = copie
  else:
    lista_libri[titolo] += copie
```

#rimuovi_libro(titolo)
Descrizione: Rimuove un libro dalla biblioteca.
Se il libro esiste, viene eliminato dal sistema.
Se il libro non esiste, il programma stampa un messaggio di errore.


```python
def rimuovi_libro(titolo):
  if titolo in lista_libri:
    del lista_libri[titolo]
  else:
    print("Errore, libro non presente")
```

#verifica_disponibilita(titolo)
Descrizione: Controlla se un libro è disponibile nella biblioteca.
Restituisce True se almeno una copia è disponibile.
Restituisce False se il libro non esiste o non ci sono copie disponibili.


```python
def verifica_disponibilità(titolo):
  if titolo in lista_libri:
    if lista_libri[titolo] >= 1:
      return True, f"copie rimanenti: {lista_libri[titolo]}"
    else:
      return False
  else:
    return False
```

#prendi_in_prestito(titolo)
Descrizione: Riduce il numero di copie disponibili di un libro di 1, simulando un prestito.
Se il libro è disponibile, decrementa il numero di copie di 1.
Se non ci sono copie disponibili o il libro non esiste, stampa un messaggio di errore.


```python
def prendi_in_prestito(titolo):
  if titolo in lista_libri:
    if lista_libri[titolo] >= 1:
      lista_libri[titolo] -= 1
      print(f"prestito registrato: {titolo}, copie rimanenti {lista_libri[titolo]}")
    else:
      print("copie insufficienti")
  else:
    print("libro non presente")
```

#statistiche_biblioteca()
Descrizione: Restituisce un dizionario con informazioni sulla biblioteca.
totale_libri: numero totale di titoli.
copie_totali: numero totale di copie.
media_copie: numero medio di copie per libro


```python
def statistiche_biblioteca():

  n_titoli = len(lista_libri)
  if n_titoli == 0:
    return{
        "n_titoli" : 0,
        "copie_totali" : 0,
        "avg_copie" : 0
    }
  copie_totali = sum(lista_libri.values())
  avg_copie = copie_totali / n_titoli

  statistiche = {
      "n_titoli" : n_titoli,
      "copie_totali" : copie_totali,
      "avg_copie" : avg_copie
  }
  return statistiche
```

#visualizza_libri()
Descrizione: Mostra un elenco di tutti i libri nella biblioteca insieme al numero di copie disponibili.
Se la biblioteca è vuota, stampa un messaggio che indica che non ci sono libri.


```python
def visualizza_libri():
  if not lista_libri:
    print("La biblioteca è vuota")
  else:
    for titolo, copie in lista_libri.items():
      print(f"è presente il seguente titolo: {titolo} con queste quantità {copie}")
visualizza_libri()
```

# restaurare_libro(titolo, copie)
Descrizione: Ripristina (aggiunge) copie di un libro che è stato preso in prestito o che necessita di più copie.
Se il libro esiste, aggiunge il numero specificato di copie.
Se il libro non esiste, stampa un messaggio di errore.



```python
def restaurare_libro(titolo,copie):
  if titolo in lista_libri:
    lista_libri[titolo] += copie
  else:
    print("il libro cercato non esiste")
```

```python
while True:
    print("\n--- BENVENUTO IN BIBLIOSOFTWARE ---")
    print("1. Aggiungi nuovo titolo o copie")
    print("2. Rimuovi un titolo dal sistema")
    print("3. Verifica disponibilità")
    print("4. Prendi in prestito ( -1 copia )")
    print("5. Restaura/Riporta copie")
    print("6. Guarda le statistiche")
    print("7. Visualizza tutti i libri")
    print("esci. Chiudi il programma")

    scelta = input("\nCosa desideri fare? ").lower()

    if scelta == "esci":
        print("Grazie per aver usato BiblioSoftware. A presto!")
        break # Ferma il ciclo e chiude il programma

    elif scelta == "1":
        t = input("Titolo del libro: ")
        c = int(input("Numero di copie: "))
        aggiungi_libro(t, c)

    elif scelta == "2":
        t = input("Titolo da rimuovere: ")
        rimuovi_libro(t)

    elif scelta == "3":
        t = input("Titolo da cercare: ")
        if verifica_disponibilità(t):
            print("Disponibile!")
        else:
            print("Non disponibile.")

    elif scelta == "4":
        t = input("Quale libro vuoi prendere in prestito? ")
        prendi_in_prestito(t)

    elif scelta == "5":
        t = input("Quale libro hai riportato o restaurato? ")
        c = int(input("Quante copie? "))
        restaurare_libro(t, c)

    elif scelta == "6":
        stats = statistiche_biblioteca()
        print("\n--- STATISTICHE ---")
        print(f"Titoli totali: {stats['n_titoli']}")
        print(f"Copie totali: {stats['copie_totali']}")
        print(f"Media copie: {stats['avg_copie']:.2f}") # .2f serve per vedere solo 2 decimali

    elif scelta == "7":
        visualizza_libri()


    else:
        print("Opzione non valida, riprova.")
```
